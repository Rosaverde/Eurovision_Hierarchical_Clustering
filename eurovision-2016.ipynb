```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/matzim95/ML-datasets/master/eurovision-2016.csv')
df = df.pivot(index='From country', columns='To country')['Jury Rank']
df.head(3)
```

```python
df.isnull().sum().sum()
```

```python
df = df.fillna(0)
```

```python
df.isnull().sum().sum()
```

### Hierarchical clustering

```python
merg = linkage(df, method='complete')

dendrogram(merg, labels = df.index, leaf_rotation= 90, leaf_font_size = 6, color_threshold=0.7*max(merg[:,2]))
plt.show()
```

```python
merg = linkage(df, method='single')

dendrogram(merg, labels = df.index, leaf_rotation= 90,  color_threshold=0.8*max(merg[:,2]))
plt.show()
```
